```python
# simple logger to track progress during training
class Logger:
  def __init__(self):
    self.losses_train = []
    self.losses_test = []
    self.accuracies_train = []
    self.accuracies_test = []

  def log(self, accuracy_train=0, loss_train=0, accuracy_test=0, loss_test=0):
    self.losses_train.append(loss_train)
    self.accuracies_train.append(accuracy_train)
    self.losses_test.append(loss_test)
    self.accuracies_test.append(accuracy_test)

  def plot_loss(self, train=True, test=True):
    if not train and not test:
      raise ValueError("Cannot plot loss because train and test")
    plt.figure()
    if train:
      plt.plot(self.losses_train, label="Train")
    if test:
      plt.plot(self.losses_test, label="Test")
    plt.legend(loc='best')
    plt.xlabel("Epoch")
    plt.title("CrossEntropy Loss")

  def plot_accuracy(self, train=True, test=True):
    if not train and not test:
      raise ValueError("Cannot plot accuracy because train and test")
    plt.figure()
    if train:
      plt.plot(self.accuracies_train, label="Train")
    if test:
      plt.plot(self.accuracies_test, label="Test")
    plt.legend(loc='best')
    plt.xlabel("Epoch")
    plt.title("Accuracy")
  
  def print_last():
    print(f"Epoch {len(self.losses_train):2d}, \
                  Train:loss={self.losses_train[-1]:.3f}, accuracy={self.accuracies_train[-1]*100:.1f}%, \
                  Test: loss={self.losses_test[-1]:.3f}, accuracy={self.accuracies_test[-1]*100:.1f}%", flush=True)
```

```python
def train(input_size, hidden_size,nb_layers, nb_labels, lr=0.01, nb_epochs=10, batch_size=64, seed=None,model_name='RNN'):
    global model
    if model_name == 'RNN':
         model = RNN(input_size, hidden_size, nb_layers, nb_labels).to(device)
    elif model_name == 'GRU':
         model = GRU(input_size, hidden_size, nb_layers, nb_labels).to(device)
    elif model_name == 'LSTM':
         model = LSTM(input_size, hidden_size, nb_layers, nb_labels).to(device)
    elif model_name == 'Transformer':
         model = Transformer(input_size, nb_labels).to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    best_model = None
    global data_loader_test
    global data_loader_train
    data_loader_train, data_loader_val, data_loader_test = get_mnist_dataloaders(0.1, batch_size=batch_size)
    
    best_val_accuracy = 0
    logger = Logger()
    import time
    start_time = time.time()


    for epoch in range(nb_epochs):
        for batch_idx, (x, y) in enumerate(data_loader_train) :
            
            x=x.to(device=device).squeeze(1)
            x=1/256*x.to(device=device).reshape(x.shape[0],sequence_length,input_size)
           
            

            y=y.to(device=device)
            y_pred = model(x)
            loss = criterion(y_pred, y)
            optimizer.zero_grad()
            loss.backward()
            #torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=2.0, norm_type=2)
            #torch.nn.utils.clip_grad_norm_(model.parameters(), 1)
            optimizer.step()         
            torch.cuda.empty_cache() 
        accuracy_train, loss_train = accuracy_and_loss_whole_dataset(batch_size,sequence_length, input_size,data_loader_train, model)
        accuracy_val, loss_val = accuracy_and_loss_whole_dataset(batch_size,sequence_length, input_size,data_loader_val, model)
        if accuracy_val  > best_val_accuracy: # select the best parameters based on the validation accuracy 
            best_model=copy.deepcopy(model)
            best_val_accuracy=accuracy_val
            pass  

        logger.log(accuracy_train, loss_train, accuracy_val, loss_val)
        if epoch % 1 == 0:
          print("--- %s seconds ---" % (time.time() - start_time))
          print(f"Epoch {epoch:2d}, \
                  Train:loss={loss_train.item():.4f}, accuracy={accuracy_train.item()*100:.4f}%, \
                  Val: loss={loss_val.item():.4f}, accuracy={accuracy_val.item()*100:.1f}%", flush=True)
        
    return best_model, best_val_accuracy, logger
```

## Vanilla RNN

```python
# create RNN

class RNN (nn.Module):
    def __init__(self,input_size, hidden_size, nb_layers, nb_labels):
        super(RNN,self).__init__()
        self.input_size = input_size
        self.nb_labels= nb_labels
        self.nb_layers = nb_layers
        self.hidden_size = hidden_size
        self.rnn = nn.RNN(input_size, hidden_size,nb_layers, batch_first=True).cuda() # RNN 
        self.fc1 = nn.Linear(hidden_size,hidden_size)
        self.fc2 = nn.Linear(hidden_size,hidden_size)
        self.fc = nn.Linear(hidden_size, nb_labels)
        
    def forward (self, x):
        h0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
        out, hn = self.rnn(x, h0) # The output shape is (minibatch, hidden_size, nb_labels) 
        out = self.fc1(hn[-1,:,:])  # The output of the last hidden state 
        out = self.fc2(out) 
        out = self.fc(out) 
        return out
    
```

```python
# gru

# create gru

class GRU (nn.Module):
    def __init__(self,input_size, hidden_size, nb_layers, nb_labels):
        super(GRU,self).__init__()
        self.input_size = input_size
        self.nb_labels= nb_labels
        self.nb_layers = nb_layers
        self.hidden_size = hidden_size
        self.gru = nn.GRU(input_size, hidden_size,nb_layers, batch_first=True) # GRU
        self.fc1 = nn.Linear(hidden_size,hidden_size)
        self.fc2 = nn.Linear(hidden_size,hidden_size)
        self.fc = nn.Linear(hidden_size, nb_labels)
        
    def forward (self, x):
        h0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
        out, _ = self.gru(x, h0)
        out = self.fc1(out[:,-1,:])   # The output of the last hidden state 
        out = self.fc2(out) 
        out = self.fc(out) 
        return out
       
    
```

```python
# LSTM

# LSTM

#class LSTM (nn.Module):
 #   def __init__(self,input_size, hidden_size, nb_layers, nb_labels):
  #      super(LSTM,self).__init__()
   #     self.input_size = input_size
    #    self.nb_labels= nb_labels
     #   self.nb_layers = nb_layers
      #  self.hidden_size = hidden_size
       # self.lstm = nn.LSTM(input_size, hidden_size,nb_layers, batch_first=True) # lstm
        #self.fc1 = nn.Linear(hidden_size,hidden_size)
        #self.fc2 = nn.Linear(hidden_size,hidden_size)
        #self.fc = nn.Linear(hidden_size, nb_labels)
        
    #def forward (self, x):
     #   h0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
      #  c0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
       # out, _ = self.lstm(x, (h0,c0))
       # out = self.fc1(out[:,-1,:])   # The output of the last hidden state 
        #out = self.fc2(out) 
        #out = self.fc(out) 
        #return out
       

class LSTM(nn.Module):
    def __init__(self,input_size, hidden_size, nb_layers, nb_labels):
        super(LSTM,self).__init__()
        self.input_size = input_size
        self.nb_labels= nb_labels
        self.nb_layers = nb_layers
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size, hidden_size,nb_layers, batch_first=True) # GRU
        self.fc1 = nn.Linear(hidden_size,hidden_size)
        self.fc2 = nn.Linear(hidden_size,hidden_size)
        self.fc = nn.Linear(hidden_size, nb_labels)
        self.norm1 = nn.LayerNorm(hidden_size) # 1st normalization
        self.norm2 = nn.LayerNorm(hidden_size)
        
    def forward (self, x):
        h0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
        c0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
        self.lstm.flatten_parameters()
        out, (hn,cn) = self.lstm(x, (h0,c0))
        out = self.fc1(self.norm1(out[:,-1,:]))   # The output of the last hidden state 
        out = self.fc2(self.norm2(out)) 
        out = self.fc(out) 
        return out
       
    
    
```

```python

class LSTM(nn.Module):
    def __init__(self,input_size, hidden_size, vocab_size, embed_size, nb_layers, nb_labels):
        super(LSTM,self).__init__()
        self.input_size = input_size
        self.nb_labels= nb_labels
        self.nb_layers = nb_layers
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(embed_size, hidden_size,nb_layers, batch_first=True) # GRU
        self.fc1 = nn.Linear(hidden_size,hidden_size)
        self.fc2 = nn.Linear(hidden_size,hidden_size)
        self.fc = nn.Linear(hidden_size, nb_labels)
        self.norm1 = nn.LayerNorm(hidden_size) # 1st normalization
        self.norm2 = nn.LayerNorm(hidden_size)
        self.word_embedding = nn.Embedding(vocab_size,embed_size)
        
    def forward (self, x):
        h0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
        c0 = torch.zeros(self.nb_layers, x.size(0), self.hidden_size).to(device)
        z=x.long() 
        input = self.word_embedding(z) # (batch_size, sequence_length, input_size) => (batch_size, sequence_length, embed_size)
        input=input.squeeze(2)
        self.lstm.flatten_parameters()
        out, (hn,cn) = self.lstm(input, (h0,c0))
        out = self.fc1(hn[-1,:,:])   # The output of the last hidden state 
        out = self.fc2(out) 
        out = self.fc(out) 
        return out
       
```

```python
class SelfAttention(nn.Module):
    def __init__(self, embed_size, heads,device):
        super(SelfAttention, self).__init__()
        self.device=device
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads
        assert(self.head_dim*heads == embed_size), "Embed size not div by heads"
        self.values = nn.Linear(self.head_dim, self.head_dim, bias = True) # The values NN
        self.keys = nn.Linear(self.head_dim,self.head_dim,bias= True) # The keys NN 
        self.queries = nn.Linear(self.head_dim,self.head_dim,bias = True)  # The query NN
        self.fc = nn.Linear(heads*self.head_dim, self.embed_size)
        
         # A fully connected NN for the output 
    def forward(self, values, keys, query, mask):
        
        N = query.shape[0]     # the batch len 
        value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]
        
        # Split embedding into multiple heads:
        values = values.reshape(N, value_len, self.heads, self.head_dim).to(self.device)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim).to(self.device)
        #query=query.reshape(N, query_len, self.heads, self.head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim).to(self.device)
        
        values= self.values(values).to(self.device)
        keys = self.keys(keys).to(self.device)
        queries = self.queries(queries).to(self.device)
        # N for N, Q for query, H for heads, K for keys, D for head_dim
        QK = torch.einsum('nqhd,nkhd -> nhqk', [queries, keys]).to(self.device)
        if mask != None :
            QK = QK.masked_fill(mask == 0, float(1e-30)).to(self.device)
        Attention = torch.softmax(QK/(self.embed_size**(1/2)), dim= 3).to(self.device)
        out = torch.einsum('nhqk, nvhd -> nqhd', [Attention, values]).to(self.device)
        out= out.reshape(N, query_len,self.head_dim*self.heads).to(self.device) # we concatenated  all the heads
        out = self.fc(out).to(self.device)
        return out 
   


# Create a transformer block
class TransformerBlock(nn.Module):
    def __init__(self, embed_size, heads, dropout, forward_expansion,device):
        super(TransformerBlock,self ).__init__()
        self.attention = SelfAttention(embed_size, heads,device)
        self.norm1 = nn.LayerNorm(embed_size) # 1st normalization
        self.norm2 = nn.LayerNorm(embed_size) # 2nd normalization
        self.feed_forward = nn.Sequential(nn.Linear(embed_size,forward_expansion*embed_size), nn.ReLU(), nn.Linear(forward_expansion*embed_size,embed_size))
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, value, key, query, mask):
        attention = self.attention( value, key, query, mask)
        x = self.dropout(self.norm1(attention + query.reshape(attention.shape) )).to(device) # 1st normalization + Skip connection + dropout
        forward = self.feed_forward(x)
        out = self.dropout(self.norm2(forward + x)) # 2nt normalization + Skip connection + dropout
        return out 

# create the encoder 
class Encoder(nn.Module):
    def __init__(self,input_size,vocab_size, embed_size, nb_layers, heads, device, forward_expansion, dropout, max_length):
        super(Encoder, self).__init__()
        self.embed_size = embed_size
        self.device = device
        self.position_embedding = nn.Embedding(sequence_length, embed_size)
        self.positional_encoding = PositionalEncoding(embed_size,dropout,max_length)
        self.layers = nn.ModuleList([TransformerBlock(embed_size, heads, dropout=dropout, forward_expansion=forward_expansion,device=device)])
        self.dropout = nn.Dropout(dropout)
        self.word_embedding = nn.Embedding(vocab_size,embed_size)
        self.norm = nn.LayerNorm(embed_size*sequence_length) 
        
    def forward(self, x, mask):
        N, sequence_length = x.shape[0], x.shape[1]
        positions = torch.arange(0, sequence_length).expand(N, sequence_length).to(self.device)
        #print(x[1])
        z=x.long()
        input = self.word_embedding(z).squeeze(2).to(self.device)# (batch_size, sequence_length, input_size) => (batch_size, sequence_length, embed_size)        
        out= self.dropout(input + self.position_embedding(positions)).to(self.device)
        #out=self.positional_encoding(input).to(self.device)
        #print(out)
        #print('a',a.shape)
        #print('input',input.shape)
        #out= self.dropout( a)
        #out= self.dropout(x + self.positional_encoding(x)).to(self.device)
        #print(out)
        # input + positional embedding
        for layer in self.layers:
            out = layer(out, out, out, mask).to(self.device)
        
        
        return out 


# create the Transformer
class Transformer(nn.Module):
    torch.cuda.empty_cache() 
    def __init__(self, input_size =input_size,vocab_size =256,embed_size =10 , target_size = nb_labels,  nb_layers=1, forward_expansion = 1, heads = 1, dropout=0.1 , device = device, max_length=784):
        super(Transformer,self).__init__()

        self.encoder = Encoder(input_size, vocab_size, embed_size,nb_layers, heads, device, forward_expansion, dropout, max_length )
        a = embed_size*sequence_length
        self.hidden_layers = []
        hidden_size = 6
        for _ in range(hidden_size):
            self.hidden_layers.append(nn.Linear(a,a))

        self.fc1 = nn.Linear(a,a)
        self.fc2 = nn.Linear(a,a)
        self.fc = nn.Linear(a, nb_labels) # Adding a  FC NN 
        self.norm1 = nn.LayerNorm(a) 
    def forward(self, x):
        src_mask = None
        trg_mask = None
        #print(x)
        
        enc_out = self.encoder(x, src_mask).reshape(x.shape[0],-1).to(device=device)
        out = self.fc1(enc_out ).to(device=device)
        for layer in self.hidden_layers:
            enc_out = layer(enc_out).to(device=device)
        #out1 = self.fc1(enc_out).to(device=device)
        #out = self.fc(out1).to(device=device)
        out = self.fc(enc_out ).to(device=device)
        return out


```

```python
class PositionalEncoding(nn.Module):

    def __init__(self, embed_size, dropout=0, max_length=sequence_length):
        super(PositionalEncoding, self).__init__()
        self.dropout = nn.Dropout(p=dropout)

        pe = torch.zeros(max_length, embed_size)
        position = torch.arange(0, max_length, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_size, 2).float() * (-math.log(10000.0) / embed_size))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:x.size(0), :]
        return self.dropout(x)
```

```python
class SelfAttention(nn.Module):
    def __init__(self, embed_size, heads):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads
        assert(self.head_dim*heads == embed_size), "Embed size not div by heads"
        self.values = nn.Linear(self.head_dim, self.head_dim, bias = False) # The values NN
        self.keys = nn.Linear(self.head_dim,self.head_dim,bias= False) # The keys NN 
        self.queries = nn.Linear(self.head_dim,self.head_dim,bias = False)  # The query NN
        self.fc = nn.Linear(heads*self.head_dim, self.embed_size)
        
         # A fully connected NN for the output 
    def forward(self, values, keys, query, mask):
        
        N = query.shape[0]     # the batch len 
        value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]
        
        # Split embedding into multiple heads:
        #print("query",query.shape)
        #print("key", keys.shape)
        #print('values',values.shape)
        #print('value_len',value_len)
        values = values.reshape(N, value_len, self.heads, self.head_dim)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim)
        #query=query.reshape(N, query_len, self.heads, self.head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim)
        
        values= self.values(values)
        keys = self.keys(keys)
        queries = self.queries(queries)
        # N for N, Q for query, H for heads, K for keys, D for head_dim
        
        QK = torch.einsum('nqhd,nkhd -> nhqk', [queries, keys])
        torch.cuda.empty_cache() 
        if mask != None :
            QK = QK.masked_fill(mask == 0, float(1e-30))
        Attention = torch.softmax(QK/(self.embed_size**(1/2)), dim= 3)
        #print('attentions',Attention.shape)
        #print('values',values.shape)

        out = torch.einsum('nhqk, nvhd -> nqhd', [Attention, values])
        out= out.reshape(N, query_len,self.head_dim*self.heads) # we concatenated  all the heads
        out = self.fc(out)
        return out 
   


# Create a transformer block
class TransformerBlock(nn.Module):
    def __init__(self, embed_size, heads, dropout, forward_expansion,device):
        super(TransformerBlock,self ).__init__()
        self.attention = SelfAttention(embed_size, heads)
        self.norm1 = nn.LayerNorm(embed_size) # 1st normalization
        self.norm2 = nn.LayerNorm(embed_size) # 2nd normalization
        self.feed_forward = nn.Sequential(nn.Linear(embed_size,forward_expansion*embed_size), nn.ReLU(), nn.Linear(forward_expansion*embed_size,embed_size))
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, value, key, query, mask):
        attention = self.attention( value, key, query, mask)
        x = self.dropout(self.norm1(attention + query.reshape(attention.shape) )).to(device) # 1st normalization + Skip connection + dropout
        forward = self.feed_forward(x)
        out = self.dropout(self.norm2(forward + x)) # 2nt normalization + Skip connection + dropout
        return out 

# create the encoder 
class Encoder(nn.Module):
    def __init__(self,src_vocab_size, embed_size, nb_layers, heads, device, forward_expansion, dropout, max_length ):
        super(Encoder, self).__init__()
        self.embed_size = embed_size
        self.device = device
        self.position_embedding = nn.Embedding(sequence_length, embed_size)
        self.positional_encoding = PositionalEncoding(embed_size,dropout,max_length)
        self.layers = nn.ModuleList([TransformerBlock(embed_size, heads, dropout=dropout, forward_expansion=forward_expansion,device=device)])
        self.dropout = nn.Dropout(dropout)
        self.word_embedding = nn.Embedding(src_vocab_size,embed_size)
        self.norm = nn.LayerNorm(embed_size*sequence_length) 
        
    def forward(self, x, mask):
        N, sequence_length = x.shape[0], x.shape[1]
        positions = torch.arange(0, sequence_length).expand(N, sequence_length).to(self.device)
        #print(x[1])
        z=x.long()
        #print('z-in',z.shape)
        input = self.word_embedding(z).squeeze(2).to(self.device)# (batch_size, sequence_length, input_size) => (batch_size, sequence_length, embed_size)    
        #print('inpu',input.shape)    
        out= self.dropout(input + self.position_embedding(positions)).to(self.device)
        #out=self.positional_encoding(input).to(self.device)
        #print(out)
        #print('a',a.shape)
        #print('input',input.shape)
        #out= self.dropout( a)
        #out= self.dropout(x + self.positional_encoding(x)).to(self.device)
        #print(out)
        # input + positional embedding
        for layer in self.layers:
            out = layer(out, out, out, mask).to(self.device)
        
        
        return out 

# create a decoder block 

class DecoderBlock(nn.Module):
    def __init__(self, embed_size, heads, forward_expansion, dropout, device):
        super(DecoderBlock, self).__init__()
        self.attention = SelfAttention(embed_size, heads)
        self.norm = nn.LayerNorm(embed_size)
        self.transformer_block = TransformerBlock(embed_size, heads, dropout, forward_expansion,device)
        self.dropout = nn.Dropout(dropout)
    def forward(self, x, value, key, src_mask, trg_mask):
        attention = self.attention(x,x,x,trg_mask)
        query = self.dropout(self.norm(attention + x))
        out = self.transformer_block(value, key, query, src_mask)
        return out 
# Create the decoder
class Decoder(nn.Module):
    def __init__(self, trg_vocab_size, embed_size, nb_layers, heads, forward_expansion, dropout, device, max_length):
        super(Decoder,self).__init__()
        self.device = device
        self.embed_size = embed_size
        self.positional_encoding = PositionalEncoding(embed_size,dropout,max_length)
        self.position_embedding = nn.Embedding(sequence_length, embed_size)
        self.dropout = nn.Dropout(dropout)
        self.word_embedding = nn.Embedding(trg_vocab_size,embed_size)
        self.layers = nn.ModuleList([DecoderBlock(embed_size, heads, forward_expansion, dropout, device) for _ in range(nb_layers)])
        self.fc = nn.Linear(embed_size, trg_vocab_size)

    def forward(self, x, enc_out, src_mask, trg_mask ):
      
        #print('src_mask',enc_out.shape)
        N, sequence_length = x.shape[0], x.shape[1]
        z=x.long()
        positions = torch.arange(0, sequence_length).expand(N, sequence_length).to(self.device)
        #print('z',z.shape)
        input = self.word_embedding(z).squeeze(2).to(self.device)# (batch_size, sequence_length, input_size) => (batch_size, sequence_length, embed_size)    
        #input = self.word_embedding(z).to(self.device)    
        #out= self.dropout(x + self.position_embedding(positions)).to(self.device)
        #x=self.positional_encoding(input).to(self.device)
        x= self.dropout(input + self.position_embedding(positions)).to(self.device)                                                      
        #print('x', x.shape)
        #print('enc_out',enc_out.shape)
        for layer in self.layers:
            x = layer(x, enc_out, enc_out, src_mask,trg_mask).to(self.device)
        out = self.fc(x)
        return out
# create a full transfromer 
class Transformer2(nn.Module):
    def __init__(self, src_vocab_size, trg_vocab_size, src_pad_idx, trg_pad_idx, embed_size=10, nb_layers=1, forward_expansion = 4, heads= 2 , dropout=0, device=device, max_length=784):
        super( Transformer2, self).__init__()
        self.encoder = Encoder(src_vocab_size,embed_size, nb_layers, heads, device, forward_expansion, dropout, max_length)
        self.decoder = Decoder(trg_vocab_size, embed_size, nb_layers, heads, forward_expansion, dropout, device, max_length )
        self.src_pad_idx = src_pad_idx
        self.trg_pad_idx = trg_pad_idx
        self.device = device
        self.fc = nn.Linear(trg_vocab_size,1)
        a = embed_size*sequence_length
        b = sequence_length*trg_vocab_size
        #print(a)
        self.fc1 = nn.Linear(b, 10)
    def make_src_mask(self, src):
        src_mask = (src != self.src_pad_idx).unsqueeze(1).unsqueeze(2).squeeze(4)
        #print(src_mask.shape)
        # (N,1,1,src_lenght)
        return src_mask.to(self.device)
    def make_trg_mask(self, trg):
        N, trg_length = trg.shape[0],trg.shape[1]
        #print (N)
        #print(trg_length)
        trg_mask = torch.tril(torch.ones((trg_length, trg_length))).expand(N,1 ,trg_length, trg_length)
        return trg_mask.to(self.device)
    def forward(self, src, trg):
        src_mask = self.make_src_mask(src)
        trg_mask = self.make_trg_mask(trg)
        #src_mask = None
        #trg_mask = None
        enc_src = self.encoder(src, src_mask)
        dec_src = self.decoder(trg, enc_src, src_mask, trg_mask )
        #print(dec_src.shape)
        out = self.fc(dec_src)
        #print(enc_src.reshape(src.shape[0],-1).shape)
        #print(a)
        #print(src.shape)
        out2 = self.fc1(dec_src.reshape(src.shape[0],-1) ).to(device=device)
        
        #print('out_shape',out.shape)
        return  out2,out




# create the Transformer
class Transformer(nn.Module):
    torch.cuda.empty_cache() 
    def __init__(self, src_vocab_size, trg_vocab_size, src_pad_idx, trg_pad_idx, embed_size=1, nb_layers=1, forward_expansion = 4, heads= 1 , dropout=0, device=device, max_length=784):
        super(Transformer,self).__init__()

        self.encoder = Encoder(src_vocab_size,embed_size, nb_layers, heads, device, forward_expansion, dropout, max_length)
        a = embed_size*sequence_length
        self.fc1 = nn.Linear(a,a)
        self.fc2 = nn.Linear(a,a)
        self.fc = nn.Linear(a, nb_labels) # Adding a  FC NN 
        self.norm1 = nn.LayerNorm(a) 
        self.src_pad_idx = src_pad_idx
        self.trg_pad_idx = trg_pad_idx
        self.device = device
        #self.fc = nn.Linear(trg_vocab_size,1)
    def make_src_mask(self, src):
        src_mask = (src != self.src_pad_idx).unsqueeze(1).unsqueeze(2).squeeze(4)
        #print(src_mask.shape)
        # (N,1,1,src_lenght)
        return src_mask.to(self.device)
    def make_trg_mask(self, trg):
        N, trg_length = trg.shape[0],trg.shape[1]
        #print (N)
        #print(trg_length)
        trg_mask = torch.tril(torch.ones((trg_length, trg_length))).expand(N,1 ,trg_length, trg_length)
        return trg_mask.to(self.device)
    def forward(self, src, trg):
        src_mask = self.make_src_mask(src)
        trg_mask = self.make_trg_mask(trg)
        #trg_mask =None
        #src_mask = None
        enc_src = self.encoder(src, src_mask)
        enc_out = self.encoder(x, src_mask).reshape(x.shape[0],-1).to(device=device)
        out = self.fc(enc_out ).to(device=device)
        return out
  

```

```python
class SelfAttention(nn.Module):
    def __init__(self, embed_size, heads):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads
        assert(self.head_dim*heads == embed_size), "Embed size not div by heads"
        self.values = nn.Linear(self.head_dim, self.head_dim, bias = False) # The values NN
        self.keys = nn.Linear(self.head_dim,self.head_dim,bias= False) # The keys NN 
        self.queries = nn.Linear(self.head_dim,self.head_dim,bias = False)  # The query NN
        self.fc = nn.Linear(heads*self.head_dim, self.embed_size)
        
         # A fully connected NN for the output 
    def forward(self, values, keys, query, mask):
        
        N = query.shape[0]     # the batch len 
        value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]
        
        # Split embedding into multiple heads:
        #print("query",query.shape)
        #print("key", keys.shape)
        #print('values',values.shape)
        #print('value_len',value_len)
        values = values.reshape(N, value_len, self.heads, self.head_dim)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim)
        #query=query.reshape(N, query_len, self.heads, self.head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim)
        
        values= self.values(values)
        keys = self.keys(keys)
        queries = self.queries(queries)
        # N for N, Q for query, H for heads, K for keys, D for head_dim
        
        QK = torch.einsum('nqhd,nkhd -> nhqk', [queries, keys])
        torch.cuda.empty_cache() 
        if mask != None :
            QK = QK.masked_fill(mask == 0, float(1e-30))
        Attention = torch.softmax(QK/(self.embed_size**(1/2)), dim= 3)
        #print('attentions',Attention.shape)
        #print('values',values.shape)

        out = torch.einsum('nhqk, nvhd -> nqhd', [Attention, values])
        out= out.reshape(N, query_len,self.head_dim*self.heads) # we concatenated  all the heads
        out = self.fc(out)
        return out 
   


# Create a transformer block
class TransformerBlock(nn.Module):
    def __init__(self, embed_size, heads, dropout, forward_expansion,device):
        super(TransformerBlock,self ).__init__()
        self.attention = SelfAttention(embed_size, heads)
        self.norm1 = nn.LayerNorm(embed_size) # 1st normalization
        self.norm2 = nn.LayerNorm(embed_size) # 2nd normalization
        self.feed_forward = nn.Sequential(nn.Linear(embed_size,forward_expansion*embed_size), nn.ReLU(), nn.Linear(forward_expansion*embed_size,embed_size))
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, value, key, query, mask):
        attention = self.attention( value, key, query, mask)
        x = self.dropout(self.norm1(attention + query.reshape(attention.shape) )).to(device) # 1st normalization + Skip connection + dropout
        forward = self.feed_forward(x)
        out = self.dropout(self.norm2(forward + x)) # 2nt normalization + Skip connection + dropout
        return out 

# create the encoder 
class Encoder(nn.Module):
    def __init__(self,src_vocab_size, embed_size, nb_layers, heads, device, forward_expansion, dropout, max_length ):
        super(Encoder, self).__init__()
        self.embed_size = embed_size
        self.device = device
        self.position_embedding = nn.Embedding(sequence_length, embed_size)
        self.positional_encoding = PositionalEncoding(embed_size,dropout,max_length)
        self.layers = nn.ModuleList([TransformerBlock(embed_size, heads, dropout=dropout, forward_expansion=forward_expansion,device=device)])
        self.dropout = nn.Dropout(dropout)
        self.word_embedding = nn.Embedding(src_vocab_size,embed_size)
        self.norm = nn.LayerNorm(embed_size*sequence_length) 
        
    def forward(self, x, mask):
       
        N, sequence_length = x.shape[0], x.shape[1]
        positions = torch.arange(0, sequence_length).expand(N, sequence_length).to(self.device) 
        out= self.dropout(x + self.position_embedding(positions)) # input + positional embedding
        for layer in self.layers:
            out = layer(out, out, out, mask)         
        return out 

# create a decoder block 

class DecoderBlock(nn.Module):
    def __init__(self, embed_size, heads, forward_expansion, dropout, device):
        super(DecoderBlock, self).__init__()
        self.attention = SelfAttention(embed_size, heads)
        self.norm = nn.LayerNorm(embed_size)
        self.transformer_block = TransformerBlock(embed_size, heads, dropout, forward_expansion,device)
        self.dropout = nn.Dropout(dropout)
    def forward(self, x, value, key, src_mask, trg_mask):
        attention = self.attention(x,x,x,trg_mask)
        query = self.dropout(self.norm(attention + x))
        out = self.transformer_block(value, key, query, src_mask)
        return out 
# Create the decoder
class Decoder(nn.Module):
    def __init__(self, trg_vocab_size, embed_size, nb_layers, heads, forward_expansion, dropout, device, max_length):
        super(Decoder,self).__init__()
        self.device = device
        self.embed_size = embed_size
        self.positional_encoding = PositionalEncoding(embed_size,dropout,max_length)
        self.position_embedding = nn.Embedding(sequence_length, embed_size)
        self.dropout = nn.Dropout(dropout)
        self.word_embedding = nn.Embedding(trg_vocab_size,embed_size)
        self.layers = nn.ModuleList([DecoderBlock(embed_size, heads, forward_expansion, dropout, device) for _ in range(nb_layers)])
        self.fc = nn.Linear(embed_size, trg_vocab_size)

    def forward(self, x, enc_out, src_mask, trg_mask ):
      
        #print('src_mask',enc_out.shape)
        N, sequence_length = x.shape[0], x.shape[1]
        #z=x.long()
        positions = torch.arange(0, sequence_length).expand(N, sequence_length).to(self.device)
        #print('z',z.shape)
        #input = self.word_embedding(z).squeeze(2).to(self.device)# (batch_size, sequence_length, input_size) => (batch_size, sequence_length, embed_size)    
        #input = self.word_embedding(z).to(self.device)    
        #out= self.dropout(x + self.position_embedding(positions)).to(self.device)
        #x=self.positional_encoding(input).to(self.device)
        x= self.dropout(x + self.position_embedding(positions)).to(self.device)
        #print('x', x.shape)
        #print('enc_out',enc_out.shape)
        for layer in self.layers:
            x = layer(x, enc_out, enc_out, src_mask,trg_mask).to(self.device)
        out = self.fc(x)
        return out
# create a full transfromer 
class Transformer2_line(nn.Module):
    def __init__(self, src_vocab_size, trg_vocab_size, src_pad_idx, trg_pad_idx, embed_size=28, nb_layers=1, forward_expansion = 4, heads= 4 , dropout=0, device=device, max_length=28):
        super( Transformer2_line, self).__init__()
        self.encoder = Encoder(src_vocab_size,embed_size, nb_layers, heads, device, forward_expansion, dropout, max_length)
        self.decoder = Decoder(trg_vocab_size, embed_size, nb_layers, heads, forward_expansion, dropout, device, max_length )
        self.src_pad_idx = src_pad_idx
        self.trg_pad_idx = trg_pad_idx
        self.device = device
        self.fc = nn.Linear(trg_vocab_size,1)
        self.fc2 = nn.Linear(784,784)
        self.fc3 = nn.Linear(784,784)
        self.fc = nn.Linear(784, nb_labels) # Adding a  FC NN 
        a = embed_size*sequence_length
        b = sequence_length*trg_vocab_size
        #print(b)
        self.fc1 = nn.Linear(b, 10)
    def make_src_mask(self, src):
        src_mask = (src != self.src_pad_idx).unsqueeze(1).unsqueeze(2).squeeze(4)
        return src_mask.to(self.device)
    def make_trg_mask(self, trg):
        N, trg_length = trg.shape[0],trg.shape[1]
        trg_mask = torch.tril(torch.ones((trg_length, trg_length))).expand(N,1 ,trg_length, trg_length)
        return trg_mask.to(self.device)
    def forward(self, src, trg):
        #src_mask = self.make_src_mask(src)
        #trg_mask = self.make_trg_mask(trg)
        src_mask=None
        trg_mask=None
        #print(src.shape)
        enc_src = self.encoder(src, src_mask)
        #print(enc_src.shape)
        dec_src = self.decoder(trg, enc_src, src_mask, trg_mask )
        #print(dec_src.shape)
        out = dec_src
        #print(enc_src.reshape(src.shape[0],-1).shape)
        #print(a)
        #print(src.shape)
        #print('dec',dec_src.shape)
        #b = sequence_length*trg_vocab_size
        x=enc_src.reshape(src.shape[0],-1)
        #print(x.shape)
        x1 = self.fc2(x ).to(device=device)
        x2= self.fc3(x1)
        out2=self.fc1(x2 + x)
        
        #print('out_shape',out.shape)
        return  out2,out
```

```python
class PositionalEncoding(nn.Module):

    def __init__(self, d_model, dropout=0.1, max_len=784):
        super(PositionalEncoding, self).__init__()
        self.dropout = nn.Dropout(p=dropout)

        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:x.size(0), :]
        return self.dropout(x)
```
